## O mundo dos dados
Vamos começar com a primeira pergunta: o que são dados? **Dados** (o plural da palavra *datum*) podem ser pensados como medições registradas de algo no mundo real. Por exemplo, uma lista de alturas é um dado – ou seja, a altura é uma medida da distância entre a cabeça de uma pessoa e seus pés. Costumamos chamar isso de algo que os dados estão descrevendo uma **unidade de observação**. No caso dessas alturas, uma pessoa é a unidade de observação.

Como você pode imaginar, há muitos dados que podemos reunir para descrever uma pessoa – incluindo idade, peso, se é fumante e muito mais. Uma ou mais dessas medições usadas para descrever uma unidade específica de observação é chamada de **ponto de dados**, e cada medição em um ponto de dados é chamada de **variável** (isso também é frequentemente chamado de recurso). Quando você tem vários pontos de dados juntos, você tem um conjunto de dados.

## Tipos de dados
Os dados também podem ser divididos em duas categorias principais: quantitativos e qualitativos:

```python
import matplotlib.pyplot as plt
from IPython.display import Image
Image("images/tipos_dados.PNG")
```

Os dados quantitativos são uma medida que pode ser descrita como um número; dados qualitativos são dados descritos por valores não numéricos, como texto. Sua altura é um dado que seria descrito como quantitativo. No entanto, descrever a si mesmo como "fumante" ou "não fumante" seria considerado um dado qualitativo.

Os dados quantitativos podem ainda ser classificados em duas subcategorias: **discretos** e **contínuos**. Valores quantitativos discretos são valores que podem assumir um nível fixo de precisão – geralmente inteiros. Por exemplo, o número de cirurgias que você fez em sua vida é um valor discreto – você pode fazer 0, 1 ou mais cirurgias, mas não pode fazer 1,5 cirurgias. Uma variável contínua é um valor que, em teoria, pode ser dividido com uma quantidade arbitrária de precisão. Por exemplo, sua massa corporal pode ser descrita com precisão arbitrária como 55, 55.3, 55.32 e assim por diante. Na prática, é claro, os instrumentos de medição limitam a nossa precisão. No entanto, se um valor puder ser descrito com maior precisão, ele será geralmente considerado contínuo.
> #### Observação
> Os dados qualitativos geralmente podem ser convertidos em dados quantitativos, e os dados quantitativos também podem ser convertidos em dados qualitativos. Isso será explicado mais adiante no módulo usando um exemplo.

Vamos pensar sobre isso usando o exemplo de ser um "fumante" versus um "não-fumante". Embora você possa se descrever na categoria de "fumante" ou "não fumante", você também pode reimaginar essas categorias como respostas à afirmação "você fuma regularmente" e, em seguida, usar os valores booleanos de 0 e 1 para representar "verdadeiro" e "falso", respectivamente.

Da mesma forma, na direção oposta, dados quantitativos, como altura, podem ser convertidos em dados qualitativos. Por exemplo, em vez de pensar na altura de um adulto como um número em polegadas ou centímetros (cm), você pode classificá-los em grupos, com pessoas maiores que 72 polegadas (ou seja, 183 cm) na categoria "alto", pessoas entre 63 polegadas e 72 polegadas (ou seja, entre 160 e 183 cm) como "médio", e pessoas com menos de 63 polegadas (ou seja, 152 cm) como "baixo".

Dados brutos, por si só, são simplesmente um grupo de valores. No entanto, não é muito interessante nesta forma. A análise de dados é o processo pelo qual convertemos dados em informação e, posteriormente, em conhecimento. Quando a análise de dados é combinada com a realização de previsões, temos a **análise de dados**.

Existem muitas ferramentas disponíveis para dar sentido aos dados. Uma das ferramentas mais poderosas na caixa de ferramentas de análise de dados é usar matemática em conjuntos de dados. Uma dessas ferramentas matemáticas é a **estatística**.

## Tipos de Estatísticas
A estatística pode ser dividida em duas subcategorias: **estatística descritiva** e a **estatística inferencial**.
Estatísticas descritivas são usadas para descrever dados. As estatísticas descritivas sobre uma única variável em um conjunto de dados são chamadas de análise **univariada**, enquanto as estatísticas descritivas que analisam duas ou mais variáveis ao mesmo tempo são chamadas de análise **multivariada**.

Em contraste, a estatística inferencial pensa em conjuntos de dados como uma **amostra** ou uma pequena porção de medições de um grupo maior chamado **população**. Por exemplo, uma pesquisa com 10.000 eleitores em uma eleição nacional é uma amostra de toda a população de eleitores de um país. A estatística inferencial é usada para tentar inferir as propriedades de uma população, com base nas propriedades de uma amostra.

### Distribuição de frequência de dados
A distribuição de dados é simplesmente uma contagem do número de valores que estão em um conjunto de dados. Por exemplo, digamos que temos um conjunto de dados de 1.000 registros médicos e uma das variáveis no conjunto de dados é a cor dos olhos. Se olharmos para o conjunto de dados e descobrirmos que 700 pessoas têm olhos castanhos, 200 pessoas têm olhos verdes e 100 pessoas têm olhos azuis, então acabamos de descrever a distribuição do conjunto de dados. Especificamente, descrevemos a **distribuição de frequência absoluta**. Se descrevêssemos as contagens não pelo número real de ocorrências no conjunto de dados, mas como a proporção do número total de pontos de dados, estaremos descrevendo sua **distribuição de frequência relativa**. No exemplo anterior de cor dos olhos, a distribuição de frequência relativa seria 70% de olhos castanhos, 20% de olhos verdes e 10% de olhos azuis.

É fácil calcular uma distribuição quando a variável pode assumir um pequeno número de valores fixos, como a cor dos olhos. Mas e quanto a uma variável quantitativa que pode assumir muitos valores diferentes, como altura? A maneira geral de calcular as distribuições para esses tipos de variáveis é criar "buckets" de intervalo aos quais esses valores podem ser atribuídos e, em seguida, calcular as distribuições usando esses buckets. Por exemplo, a altura pode ser dividida em baldes de intervalo de 5 cm para fazer a seguinte distribuição absoluta. Podemos então dividir cada linha na tabela pelo número total de pontos de dados (ou seja, 10.000) e obter a distribuição relativa.

Outra coisa útil a fazer com distribuições é representá-las graficamente. Vamos agora criar um **histograma**, que é uma representação gráfica da distribuição contínua usando *buckets* de intervalo.

## Quantil
Uma maneira de quantificar a distribuição de dados numericamente é usar quantis. *N-quantis* são um conjunto de *n-1* pontos usados para dividir uma variável em n grupos. Esses pontos são frequentemente chamados de **pontos de corte**. Por exemplo, um 4-quantil (também conhecido como quartis) é um grupo de três pontos que dividem uma variável em quatro grupos de números aproximadamente iguais. Existem vários nomes comuns para quantis que são usados de forma intercambiável, e são os seguintes:

```python
Image("images/quantiles.PNG")
```

O procedimento para calcular os quantis varia de lugar para lugar. Usaremos o seguinte procedimento para calcular os `n-quantis` para `d` pontos de dados para uma única variável:
1. Ordene os pontos de dados do menor para o maior.
2. Determine o número `n` de `n-quantis` que deseja calcular e o número de pontos de corte, *n-1*.
3. Determine qual número *k* ponto de corte você deseja calcular, ou seja, um número de 1 a n-1. Se você estiver iniciando o cálculo, defina *k* igual a 1.
4. Encontre o índice, *i*, para o *k-ésimo* ponto de corte usando a seguinte equação:

$$
\small i = \left [ \dfrac{k}{n}(d-1) \right ] + 1
$$

5. Se `i` calculou no número **3** é um número inteiro, basta escolher esse item numerado dos pontos de dados ordenados. Se o **k-ésimo** ponto de corte não for um número inteiro, encontre o item numerado que é menor que `i` e o que vem depois dele. Multiplique a diferença entre o item numerado e o que vem depois dele e, em seguida, multiplique pela parte decimal do índice. Adicione este número ao item de menor número.
6. Repita os passos 2 a 5 com diferentes valores de `k` até ter calculado todos os pontos de corte.

Tendencia central
Uma das perguntas comuns feitas a uma variável em um conjunto de dados é qual é um valor típico para essa variável. Este valor é frequentemente descrito como a tendência central da variável. Existem muitos números calculados a partir de um conjunto de dados que são frequentemente usados para descrever sua tendência central, cada um com suas próprias vantagens e desvantagens. Algumas das maneiras de medir a tendência central incluem o seguinte:
* **Moda**: A moda é simplesmente o valor que aparece com mais frequência na distribuição de uma variável. Na Figura acima, o exemplo da cor dos olhos, a moda seria "olhos castanhos", porque ocorre com mais frequência no conjunto de dados. Se vários valores estiverem vinculados à variável mais comum, a variável será chamada de **multimodal** e todos os valores mais altos serão relatados. Se nenhum valor for repetido, não haverá moda para esses conjuntos de valores. A moda tende a ser útil quando uma variável pode assumir um número pequeno e fixo de valores. No entanto, é problemático calcular quando uma variável é uma variável quantitativa contínua, como em nosso problema de altura. Com essas variáveis, outros cálculos são mais adequados para determinar a tendência central.
* **Média**: A média de uma variável é o valor calculado quando você soma todos os valores da variável e divide pelo número de pontos de dados. Por exemplo, digamos que você tenha um pequeno conjunto de dados de idades: 26, 25, 31, 35 e 29. A média dessas idades seria 29,2, porque esse é o número que você obtém ao somar os 5 números e depois dividir por 5, ou seja, o número de pontos de dados. A média é fácil de calcular e geralmente faz um bom trabalho ao descrever um valor "típico" para uma variável. Não à toa, é uma das estatísticas descritivas mais relatadas na literatura. A média como tendência central, no entanto, sofre de uma grande desvantagem – é sensível a **valores discrepantes** (*Outliers*). Outliers são dados que são significativamente diferentes em valor do restante dos dados e ocorrem muito raramente. Os *outliers* geralmente podem ser identificados usando técnicas gráficas (como gráficos de dispersão e gráficos de caixa) e identificando quaisquer pontos de dados que estejam muito distantes do restante dos dados. Quando um conjunto de dados tem um *outlier*, ele é chamado de conjunto de **dados assimétrico**. Alguns motivos comuns pelos quais ocorrem discrepâncias incluem dados impuros, eventos extremamente raros e problemas com instrumentos de medição. Os valores atípicos geralmente distorcem a média a um ponto em que não são mais representativos de um valor típico nos dados.

* **Mediana**: A mediana (também chamada de segundo quartil e quinquagésimo percentil) é uma medida estranha de tendência central, mas tem algumas vantagens sérias em comparação com a média. Para calcular a mediana, pegue os números de uma variável e classifique do menor para o maior e, em seguida, determine o número do meio. Para um número ímpar de pontos de dados, esse número é simplesmente o valor médio dos dados ordenados. Se houver um número par de pontos de dados, tire a média dos dois números do meio.
Embora a mediana seja um pouco difícil de calcular, ela é menos afetada por valores discrepantes, ao contrário da média. Para ilustrar esse fato, calcularemos a mediana do conjunto de dados de idade assimétrica de 26, 25, 31, 35, 29 e 82. Desta vez, quando calculamos a mediana do conjunto de dados, obtemos o valor de 30. Esse valor é muito mais próximo do valor típico do conjunto de dados do que a média de 38. Essa robustez em relação a valores discrepantes é uma das principais razões pelas quais uma mediana é calculada.

Como regra geral, é uma boa ideia calcular a média e a mediana de uma variável. Se houver uma diferença significativa no valor da média e da mediana, o conjunto de dados pode ter **outliers**.

## Dispersão
Outra propriedade de interesse em um conjunto de dados é descobrir o quão próximos os pontos de dados estão em uma variável. Por exemplo, os conjuntos de números [100, 100, 100] e [50, 100, 150] têm uma média de 100, mas os números do segundo grupo estão mais espalhados do que o primeiro. Essa propriedade de descrever como os dados são espalhados é chamada de **dispersão**.

Há muitas maneiras de medir a dispersão de uma variável. Aqui estão algumas das maneiras mais comuns de avaliar a dispersão:
* **Faixa**: A faixa é simplesmente a diferença entre os valores mais alto e mais baixo de uma variável. É incrivelmente fácil de calcular, mas é muito suscetível a valores discrepantes. Também não fornece muitas informações sobre a dispersão de valores no meio do conjunto de dados.
* **Desvio padrão/variância**: O desvio padrão é simplesmente a raiz quadrada da média da diferença quadrada entre cada ponto de dados e a média. O valor do desvio padrão varia de 0 até infinito positivo. Quanto mais próximo o desvio padrão estiver de 0, menor será a variação dos números no conjunto de dados. Se o desvio padrão for 0, isso significa que todos os valores de uma variável do conjunto de dados são os mesmos.

    Uma distinção sutil a ser observada é que existem duas fórmulas diferentes para o desvio padrão, que são mostradas abaixo. Quando o conjunto de dados representa a população inteira, você deve calcular o desvio padrão da população usando a fórmula A na Fórmula abaixo. Se sua amostra representa uma parte das observações, você deve usar a fórmula B para o desvio padrão da amostra, conforme mostrado abaixo. Na dúvida, use a variância amostral, pois ela é considerada mais conservadora. Além disso, na prática, a diferença entre as duas fórmulas é muito pequena quando há muitos pontos de dados.
    O desvio padrão é geralmente a quantidade usada com mais frequência para descrever a dispersão. No entanto, como o alcance, ele também pode ser afetado por valores discrepantes, embora não tão extremamente quanto o alcance. Também pode ser bastante envolvido para calcular. Ferramentas modernas, no entanto, geralmente facilitam o cálculo do desvio padrão. Uma observação final é que, ocasionalmente, você pode ver um valor relacionado, variância, listado também. Esta quantidade é simplesmente o quadrado do desvio padrão:

$$
\small A)\quad \sqrt{\dfrac{\sum^n_{i=1}(x_i - u_x)^2}{n}} \quad B)\quad \sqrt{\dfrac{\sum^n_{i=1}(x_i - u_x)^2}{n-1}}
$$

* **Intervalo Interquartil (IQR)**: O intervalo interquartil é a diferença entre o primeiro quartil, Q1 (também chamado de quartil inferior), e o terceiro quartil, Q3 (também chamado de quartil superior).

> #### Observação
> Para obter mais informações sobre o cálculo de quantis e quartis, consulte a seção *Distribuição de dados* neste módulo.

O IQR, ao contrário do intervalo e do desvio padrão, é robusto em relação a valores *outliers* e, portanto, embora seja a função mais complicada de calcular, fornece uma maneira mais robusta de medir a dispersão dos conjuntos de dados. Na verdade, o IQR é frequentemente usado para definir valores discrepantes. Se um valor em um conjunto de dados for menor que Q1 - 1.5 X IQR ou maior que Q3 + 1.5 X IQR, o valor será considerado um valor discrepante.

Uma das tendências mais comuns na análise de dados bivariados são as tendências lineares. > #### Observação
> Para obter mais informações sobre regressão OLS, consulte um livro de estatística, como Estatísticas de David Freedman, Robert Pisani e Roger Purves. Um método para quantificar a correlação linear é usar o que é chamado de coeficiente de correlação de Pearson. O coeficiente de correlação de Pearson, geralmente representado pela letra `r`, é um número que varia de -1 a 1, indicando quão bem um gráfico de dispersão se ajusta a uma tendência linear. Para calcular o coeficiente de correlação de Pearson, r, usamos a seguinte fórmula:
$$
\small r = \dfrac{\sum^n_{i=1}(x_i - \bar{x})(y - \bar{y})}{\sqrt{\sum^n_{i=1}(x_i - \bar{x})^2}\:\sqrt{\sum^n_{i=1}(y_i - \bar{y})^2}}
$$


Essa fórmula é um pouco pesada, então vamos trabalhar com um exemplo para transformar a fórmula em etapas específicas.

Finalmente, um dos principais problemas associados ao cálculo da correlação é a falácia lógica da correlação que implica causalidade. Ou seja, só porque x e y têm uma forte correlação não significa que x causa y. Vejamos nosso exemplo do número de horas trabalhadas versus o número de vendas líquidas por semana. Como profissional de análise, você será responsável por evitar armadilhas, como correlação e causalidade, e pensar criticamente sobre todas as possibilidades que podem ser responsáveis ​​pelos resultados que você vê.

Um dos muitos problemas com os quais você pode ter que lidar ao trabalhar com conjuntos de dados é a falta de valores. No entanto, nesta seção, gostaríamos de dedicar algum tempo para discutir algumas das estratégias que você pode usar para lidar com dados ausentes. Algumas de suas opções incluem o seguinte:

* **Excluindo linhas**: se um número muito pequeno de linhas (ou seja, menos de 5% do seu conjunto de dados) estiver com dados ausentes, a solução mais simples pode ser apenas excluir os pontos de dados do seu conjunto. Tal resultado não deve afetar excessivamente seus resultados. 
* **Imputação de média/mediana/modo**: Se 5% a 25% de seus dados para uma variável estiverem faltando, outra opção é pegar a média, mediana ou moda dessa coluna e preencher os espaços em branco com esse valor. Isso pode fornecer um pequeno viés para seus cálculos, mas permitirá que você conclua mais análises sem excluir dados valiosos.
* **Imputação de regressão**: Se possível, você pode construir e usar um modelo para imputar valores ausentes. Essa habilidade pode estar além da capacidade da maioria dos analistas de dados, mas se você estiver trabalhando com um cientista de dados, essa opção pode ser viável.
* **Exclusão de variáveis**: Em última análise, você não pode analisar dados que não existem. Se você não tiver muitos dados disponíveis e uma variável estiver com a maioria dos dados faltando, pode ser simplesmente melhor remover essa variável do que fazer muitas suposições e chegar a conclusões erradas.

Você também descobrirá que uma parte decente da análise de dados é mais arte do que ciência. Trabalhar com dados ausentes é uma dessas áreas. Com a experiência, você encontrará uma combinação de estratégias que funcionam bem para diferentes cenários.

### Teste de significância estatística
Outra parte da análise que é útil na análise de dados é o teste de significância estatística. Muitas vezes, um analista está interessado em comparar as propriedades estatísticas de dois grupos, ou talvez apenas um grupo antes e depois de uma mudança. Claro, a diferença entre esses dois grupos pode ser apenas devido ao acaso. Um exemplo de onde isso acontece é nos testes **A/B de marketing**. As empresas costumam testar dois tipos diferentes de páginas de destino para um produto e medir a **taxa de cliques (CTR)**. Você pode descobrir que a CTR da variação A da página de destino é de 10% e a CTR da variação B é de 11%. Então, isso significa que a variação B é 10% melhor que A, ou isso é apenas resultado da variação do dia-a-dia? Testes estatísticos nos ajudam a determinar exatamente isso.

Em testes estatísticos, há algumas partes importantes que você precisa ter (Figura 1.32). Primeiro, temos a estatística de teste que estamos examinando. Pode ser uma proporção, uma média, a diferença entre dois grupos ou uma distribuição. A próxima parte necessária é uma hipótese nula, que é a ideia de que os resultados observados são produto do acaso. Você precisará então de uma hipótese alternativa, que é a ideia de que os resultados vistos não podem ser explicados apenas por acaso. Finalmente, um teste precisa de um nível de significância, que é o valor que a estatística de teste precisa tomar antes de ser decidido que a hipótese nula não pode explicar a diferença. Todos os testes de significância estatística têm esses quatro aspectos, e é simplesmente uma questão de como esses componentes são calculados que diferenciam os testes de significância:
